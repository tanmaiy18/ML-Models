```python
import pandas as pd
import numpy as np
import seaborn as sns
import time,sys,os,warnings
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score , confusion_matrix, ConfusionMatrixDisplay
print("Done Importing")
```

```python
warnings.filterwarnings("ignore")

pd.set_option("display.max_rows",500)
pd.set_option("display.max_columns",500)
df=pd.read_csv("../input/avoidoverfitting/train.csv")
```

```python
def xandy(frame):
    inden =frame.drop(["target"],axis=1)
    target_var = frame["target"]
    return inden,target_var
```

```python
df.corr()
```

```python
import scipy.stats as stats
#  depen = Categorical, inden = Continous ==> pointbiserial test
features = []
alpha = 0.05
for col in df.columns:
    result, pvalue = stats.pointbiserialr(df["target"], df[col])
    
    if pvalue>alpha:
        print(f"{col}, No realtion")
    else:
        features.append(col)

```

```python
df = df[features]
```

```python
df.shape
df.corr()
```

```python
x_df,y_df=xandy(df)
```

```python
from imblearn.over_sampling import RandomOverSampler

over=RandomOverSampler()
over_x,over_y=over.fit_sample(x_df,y_df)

x_train1,x_test1=train_test_split(over_x,test_size=0.1,random_state=12)
y_train1,y_test1=train_test_split(over_y,test_size=0.1,random_state=12)
```

```python

from sklearn.metrics import roc_auc_score
model=LogisticRegression()
model.fit(x_train1,y_train1)


pred_test=model.predict(x_test1)
pred_train=model.predict(x_train1)

score_test=roc_auc_score(y_test1,pred_test)
print("Testing accuracy",score_test)

score_train=roc_auc_score(y_train1,pred_train)
print("Training Accuracy",score_train)

score_matrix =confusion_matrix(y_test1,pred_test)
print(score_matrix)
```
